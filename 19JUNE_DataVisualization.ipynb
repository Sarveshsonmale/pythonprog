```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([23, 34, 42, 49, 55, 61, 68, 73, 78, 82])
y = np.array([25, 29, 33, 37, 44, 47, 57, 62, 67, 71])
z = np.array([12, 81, 86, 88, 72, 77, 90, 92, 94, 98])
```

### Line Plot

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(1, 90, 5))
plt.xticks(range(1, 90, 5))
plt.plot(x,y, marker='o',mfc = 'red' , color = 'black' ,ls = '--', ms=5, label = '2023');
plt.plot(y,z, marker='o',mfc = 'green' , color = 'black' ,ls = '--', ms=5, label = '2024');
plt.grid()
plt.legend()

# markers => >, <, ^, H, o, *
```

## Bar Plot

```python
courses = ['java', 'python', 'koltin', 'dart', 'R']
students = [65 , 70 , 59 , 55 , 47]
```

## Box Plot

```python



marks = [67, 65, 55, 59, 69, 72, 51, 70, 53, 98, 23]
```

```python
plt.boxplot(marks);
```

```python
q3 = np.percentile(marks, 75)
q1 = np.percentile(marks, 25)
q3,q1
```

```python
iqr = q3 - q1
iqr
```

```python
ub = q3 + (1.5 * iqr)
lb = q1 - (1.5 * iqr)
```
